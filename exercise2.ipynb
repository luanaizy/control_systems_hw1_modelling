```python
import sympy as sym
from sympy.abc import s,t
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = 31/9 - (13/9)*sym.exp(-3*t)-(4/3)*t*sym.exp(-3*t)
```

```python
y_func = sym.lambdify(t, y, modules=['numpy'])
t_vals = np.linspace(0, 5, 400)
y_vals = y_func(t_vals)

plt.figure(figsize=(8, 4))
plt.plot(t_vals, y_vals, label='y(t)', color='blue')
plt.title('Resposta ao impulso do Sistema')
plt.xlabel('Tempo (t)')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
